```python
import os
os.environ['CUDA_LAUNCH_BLOCKING'] = '1'
import numpy as np
import torch
import torch.nn as nn
from torch.optim import Adam
from torch.optim.lr_scheduler import ReduceLROnPlateau
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split,TensorDataset
```

```python
# Config
INPUT_DIM = 23
NUM_CLASSES = 10
H1, H2, H3 = 128, 64, 32
HEAD_HIDDEN = 16
DROPOUT_P = 0.3
BATCH_SIZE = 64
EPOCHS = 75
LR = 1e-4
VAL_SPLIT = 0.2
CHECKPOINT_PATH = "best_multitask_model.pt"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
SEED = 45
torch.manual_seed(SEED)
np.random.seed(SEED)
```

```python
# Dataset
class ProductivityDataset(Dataset):

    def __init__(self, X, y_prod, y_stress, y_job):
        if isinstance(X, np.ndarray):
            X = torch.from_numpy(X.astype(np.float32))
        if isinstance(y_prod, np.ndarray):
            y_prod = torch.from_numpy(y_prod.astype(np.int64))
        if isinstance(y_stress, np.ndarray):
            y_stress = torch.from_numpy(y_stress.astype(np.int64))
        if isinstance(y_job, np.ndarray):
            y_job = torch.from_numpy(y_job.astype(np.int64))

        self.X = X.float()
        self.y_prod = y_prod.long()
        self.y_stress = y_stress.long()
        self.y_job = y_job.long()

        assert self.X.ndim == 2 and self.X.shape[1] == INPUT_DIM
        n = len(self.X)
        assert len(self.y_prod) == n and len(self.y_stress) == n and len(self.y_job) == n

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return (
            self.X[idx],
            self.y_prod[idx],
            self.y_stress[idx],
            self.y_job[idx],
        )
```

```python
#  Model (shared trunk + 3 heads)
class MultiTaskNet(nn.Module):
    def __init__(self, input_dim=INPUT_DIM, hidden1=H1, hidden2=H2, hidden3=H3,
                 head_hidden=HEAD_HIDDEN, num_classes=NUM_CLASSES, dropout_p=DROPOUT_P):
        super().__init__()
        self.shared = nn.Sequential(
            nn.Linear(input_dim, hidden1),
            nn.ReLU(),
            nn.Linear(hidden1, hidden2),
            nn.ReLU(),
            nn.Dropout(dropout_p),
            nn.Linear(hidden2, hidden3),
            nn.ReLU(),
        )
        # Productivity head (Regression)
        self.prod_head = nn.Sequential(
            nn.Linear(hidden3, head_hidden),
            nn.ReLU(),
            nn.Linear(head_hidden, 1)  # Single output for regression
        )
        # Stress head (Classification)
        self.stress_head = nn.Sequential(
            nn.Linear(hidden3, head_hidden),
            nn.ReLU(),
            nn.Linear(head_hidden, num_classes) # logits (no softmax here)
        )
        # Job satisfaction head (Classification)
        self.job_head = nn.Sequential(
            nn.Linear(hidden3, head_hidden),
            nn.ReLU(),
            nn.Linear(head_hidden, num_classes)
        )

    def forward(self, x):
        z = self.shared(x)
        out_prod = self.prod_head(z)
        out_stress = self.stress_head(z)
        out_job = self.job_head(z)
        return out_prod, out_stress, out_job # logits for classification, single value for regression
```

```python
#  Train / Eval utilities
def accuracy_from_logits(logits, y_true):
    preds = torch.argmax(logits, dim=1)
    return (preds == y_true).float().mean().item()

def run_epoch(model, loader, optimizer=None, device=DEVICE):
    training = optimizer is not None
    model.train(mode=training)

    ce = nn.CrossEntropyLoss()
    mse = nn.MSELoss(reduction='mean')  # or L1Loss for MAE
    total_loss = 0.0
    total_batches = 0
    acc_s_sum = 0.0
    acc_j_sum = 0.0
    prod_mse_sum = 0.0
    total_samples = 0

    for X, yP, yS, yJ in loader:
        X = X.to(device)
        # PROD should be float for regression
        yP = yP.to(device).float()
        yS = yS.to(device).long()
        yJ = yJ.to(device).long()

        if training:
            optimizer.zero_grad()

        outP, outS, outJ = model(X)        # outP: [B,1], outS: [B,C], outJ: [B,C]
        outP = outP.squeeze(1)             # [B]

        lossP = mse(outP, yP)              # scalar
        lossS = ce(outS, yS)
        lossJ = ce(outJ, yJ)

        # Use weighted sum (weights sum to 1.0)  do NOT divide further
        loss = 0.6 * lossP + 0.25 * lossS + 0.15 * lossJ

        if training:
            loss.backward()
            optimizer.step()

        bsz = X.size(0)
        total_loss += loss.item() * bsz
        total_batches += 1
        total_samples += bsz
        prod_mse_sum += lossP.item() * bsz
        acc_s_sum += accuracy_from_logits(outS, yS) * bsz
        acc_j_sum += accuracy_from_logits(outJ, yJ) * bsz

    avg_loss = total_loss / total_samples
    avg_prod_mse = prod_mse_sum / total_samples
    avg_acc_s = acc_s_sum / total_samples
    avg_acc_j = acc_j_sum / total_samples

    return avg_loss, avg_prod_mse, avg_acc_s, avg_acc_j
 # Return average productivity loss instead of accuracy
```

```python
def fit_model(train_ds, val_ds, epochs=EPOCHS, lr=LR, device=DEVICE, checkpoint_path=CHECKPOINT_PATH):
    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=2)
    val_loader = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)

    model = MultiTaskNet().to(device)
    optimizer = optim.Adam(model.parameters(), lr=lr)

    best_val_loss = float("inf")
    for epoch in range(1, epochs + 1):
        tr_loss, tr_prod_mse, tr_acc_s, tr_acc_j = run_epoch(model, train_loader, optimizer=optimizer, device=device)
        # eval (optimizer=None)
        vl_loss, vl_prod_mse, vl_acc_s, vl_acc_j = run_epoch(model, val_loader, optimizer=None, device=device)

        print(f"Epoch {epoch}/{epochs} | Train Loss: {tr_loss:.4f} | Val Loss: {vl_loss:.4f} | "
              f"Val Prod MSE: {vl_prod_mse:.4f}, Stress Acc: {vl_acc_s:.3f}, Job Acc: {vl_acc_j:.3f}")

        if vl_loss < best_val_loss:
            best_val_loss = vl_loss
            torch.save(
                {
                    "model_state": model.state_dict(),
                    "best_val_loss": best_val_loss,
                    "config": {
                        "input_dim": INPUT_DIM,
                        "num_classes": NUM_CLASSES,
                        "hidden": [H1, H2, H3],
                        "head_hidden": HEAD_HIDDEN,
                        "dropout": DROPOUT_P,
                    },
                },
                checkpoint_path,
            )

    print(f"Best val loss: {best_val_loss:.4f} (checkpoint saved to '{checkpoint_path}')")
    return model

if __name__ == "__main__":

    train_ds = TensorDataset(X_train, y_prod_train, y_stress_train, y_job_train)
    val_ds   = TensorDataset(X_val, y_prod_val, y_stress_val, y_job_val)
    test_ds =TensorDataset(X_test,y_prod_test, y_stress_test, y_job_test)
    test_loader = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False, drop_last=False)


    # Train
    model = fit_model(train_ds, val_ds, epochs=EPOCHS, lr=LR, device=DEVICE, checkpoint_path=CHECKPOINT_PATH)

    # Load best checkpoint (optional demo)
    if os.path.exists(CHECKPOINT_PATH):
        ckpt = torch.load(CHECKPOINT_PATH, map_location=DEVICE)
        best_model = MultiTaskNet().to(DEVICE)
        best_model.load_state_dict(ckpt["model_state"])
        best_model.eval()

        # Inference on a small batch
        X_batch = torch.from_numpy(X[:5].astype(np.float32)).to(DEVICE)
        with torch.no_grad():
            logits_prod, logits_stress, logits_job = best_model(X_batch)
            probs_prod = torch.softmax(logits_prod, dim=1)
            probs_stress = torch.softmax(logits_stress, dim=1)
            probs_job = torch.softmax(logits_job, dim=1)

        print("\nSample predictions (class probs sum to 1 per head):")
        print("Productivity probs:\n", probs_prod.cpu().numpy())
        print("Stress probs:\n", probs_stress.cpu().numpy())
        print("Job Satisfaction probs:\n", probs_job.cpu().numpy())
```
